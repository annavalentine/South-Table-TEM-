```python
data = pd.read_csv("TEM_data.csv")
data = data.drop([30])
data = data.drop(columns=['Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20','Unnamed: 21','Unnamed: 22','Unnamed: 23','Unnamed: 24','Unnamed: 25','Unnamed: 26','Unnamed: 27','Unnamed: 28','Unnamed: 29','Unnamed: 30'])

data = data.drop([1,3,8])

data

## I do still have the wrong elevation for station 17, sorry.
```

```python
plt.figure()
plt.scatter(data['Easting'],data['Northing'],c=data['Elevation (m from DEM)'])
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.colorbar(label="Elevation (m)")
for i in data.index:
    plt.text(x=data['Easting'][i]+0.3,y=data['Northing'][i]+0.3,s=data['Project-Station'][i])
plt.show()

vmin = 15
vmax = 65

plt.figure()
plt.title("Tied Inversion (East to West)")
# plt.plot(data['Easting'],data['Northing'])
plt.scatter(data['Easting'],data['Northing'],c=data['Tied Layer Depth (EW)'],vmin=vmin,vmax=vmax)
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.colorbar(label="Elevation (m)")
plt.show()

plt.figure()
plt.title("Point Inversion")
plt.scatter(data['Easting'],data['Northing'],c=data['Layer Depth'],vmin=vmin,vmax=vmax)
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.colorbar(label="Elevation (m)")
plt.show()

plt.figure()
plt.title("Tied Inversion (Nearest Neighbor)")
plt.scatter(data['Easting'],data['Northing'],c=data['Tied Layer Depth (NN)'],vmin=vmin,vmax=vmax)
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.colorbar(label="Elevation (m)")
plt.show()
```

```python
## Diagonal transect we decided on for processing.
transect1 = data[data['Project-Station'].isin(['21-1','29-1','24-1','25-1','26-1','27-1','41-1','43-1'])]
transect1 = transect1.sort_values("Easting")
transect1

## N-S transect 1 (W)
transect2 = data[data['Project-Station'].isin(['43-1','42-1','41-1','27-1','26-1','25-1','43-3','43-5','43-6','43-8'])]
transect2 = transect2.sort_values("Northing")

## E-W transect (...diagonal)
transect3 = data[data['Project-Station'].isin(['44,1','45-1','46-1','48-1','26-1','25-1','24-1','17-3'])]
transect3 = transect3.sort_values("Easting")

## N-S transect 2 (E)
transect4 = data[data['Project-Station'].isin(['29-1','21-1','20-1','31-1','22-1','32-1','34-1'])]
transect4 = transect4.sort_values("Northing")
```

```python
transect = transect1

## Map
plt.figure()
plt.plot(transect['Easting'],transect['Northing'])
plt.scatter(data['Easting'],data['Northing'],c=data['Elevation (m from DEM)'])
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.colorbar(label="Elevation (m)")
for i in transect.index:
    plt.text(x=transect['Easting'][i]+0.3,y=transect['Northing'][i]+0.3,s=transect['Project-Station'][i])
plt.show()

## Cross section
plt.figure()
plt.plot(transect['Easting'],transect['Elevation (m from DEM)'],label='Ground Surface')
plt.plot(transect['Easting'],transect['Elevation (from ABEM)'],label='Ground Surface (ABEM)')
plt.plot(transect['Easting'],transect['Elevation (m from DEM)']-transect['Tied Layer Depth'],label="Tied Inversion")
plt.plot(transect['Easting'],transect['Elevation (m from DEM)']-transect['Layer Depth'],label="Point Inversion")
plt.xlabel("Easting (m)")
plt.ylabel("Elevation (m)")
plt.legend()
plt.show()

plt.figure()
plt.plot(transect['Northing'],transect['Elevation (m from DEM)'],label='Ground Surface')
plt.plot(transect['Northing'],transect['Elevation (from ABEM)'],label='Ground Surface (ABEM)')
plt.plot(transect['Northing'],transect['Elevation (m from DEM)']-transect['Tied Layer Depth'],label="Tied Inversion")
plt.plot(transect['Northing'],transect['Elevation (m from DEM)']-transect['Layer Depth'],label="Point Inversion")
plt.xlabel("Easting (m)")
plt.ylabel("Elevation (m)")
plt.xlim(max(transect['Northing'])+50, min(transect['Northing']-50))
plt.legend()
plt.show()
```
